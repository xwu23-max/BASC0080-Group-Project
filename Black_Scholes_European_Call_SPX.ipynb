```python
import math
from scipy.stats import norm


def black_scholes_price(S0, K, T, r, sigma, q):
    """
    Black-Scholes-Merton model for European call and put options.

    Parameters
    ----------
    S0 : float
        Current price of the underlying asset
    K : float
        Strike price
    T : float
        Time to maturity in years
    r : float
        Risk-free interest rate (annual, continuously compounded, decimal form)
    sigma : float
        Volatility of underlying asset (annualised, decimal form)
    q : float, optional
        Continuous dividend yield (annual, decimal form), default = 0.0

    Returns
    -------
    dict
        Dictionary containing d1, d2, call_price, put_price
    """

    d1 = (math.log(S0 / K) + (r - q + 0.5 * sigma**2) * T) / (sigma * math.sqrt(T))
    d2 = d1 - sigma * math.sqrt(T)

    call_price = S0 * math.exp(-q * T) * norm.cdf(d1) - K * math.exp(-r * T) * norm.cdf(d2)
    put_price = K * math.exp(-r * T) * norm.cdf(-d2) - S0 * math.exp(-q * T) * norm.cdf(-d1)

    return {
        "d1": d1,
        "d2": d2,
        "call_price": call_price,
        "put_price": put_price
    }


# Example usage
if __name__ == "__main__":
    # Example inputs
    S0 = 6672.62      # current stock price 12 Mar 2026
    K = 6675       # strike price for a near at-the-money option
    T = 0.5       # half year to maturity
    r = 0.0354      # using the latest official 6-month Treasury bill rate from the Federal Reserve H.15 release for 12 March 2026
    sigma = 0.181223169  # annualised volatility
    q = 0.011      # based on State Street’s published SPY yield figures

    result = black_scholes_price(S0, K, T, r, sigma, q)

    print("d1 =", round(result["d1"], 6))
    print("d2 =", round(result["d2"], 6))
    print("Call Price =", round(result["call_price"], 6))
```
